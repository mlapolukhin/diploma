```python
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x):
    return 1 / (1 + np.e ** (-x))

def activation(x, m, k, h):
    return sum([
        sigmoid(k * (x + h * (n - (m + 1) / 2)))
        for n in range(1, m + 1)
    ])


x_arr = np.arange(-1.5, 1.5, 0.01)
m = 3
k = 10
h = 1
y_arr = [activation(x, m, k, h) for x in x_arr]

plt.plot(x_arr, y_arr);
```
